```python
# Функція, яка буде збирати атрибути з івентів
def attrib_to_feach(x, event="Add Vehicle Success", att="Model"):
    x = x[x["event_name"] == event]["event_attributes"].values
    new_colum = []
    for row in x:
        row = json.loads(row)
        try:
            new_colum.append(row[att])
        except:
            pass
    return new_colum
```

```python
feach_dict = {
    'Add Vehicle Success': ['Year', 'Model', 'Temporary', 'Make', 'Start Date', 'End Date'], ### diff
    'Add Payment Method Success': ['Tokenized Pay', 'Payment Method'],
    'Sign Up Success': ['Method'],
    'Add Payment Method Failed': ['Error Code'],
    'Transaction Refund': ['Amount'], ### summ
    'Subscription Premium': ['Start Date', 'End Date', 'Membership Type'], ### diff
    'Chat Conversation Opened': ['From', 'Transaction type'],
    'Account Setup Profile Skip': ['Screen'],
    'Account History Transaction Details': ['Navigated From', 'Status', 'Type'],
    'Sign Out': ['Type'],
    'Order': ['Premium Membership'],
    'Sign Up Error': ['Method'],
    'Subscription Premium Renew':  ['Start Date', 'End Date', 'Membership Type'], ### diff
    'Add Vehicle Failed': ['Error Code']
}
```

```python
data_user_3 = pd.DataFrame()
for feach in feach_dict:
    for attrib in feach_dict[feach]:
        name = feach.replace(" ", "_") + "_" + attrib.replace(" ", "_")
        data_user_3[name] = data_2.groupby("userid").apply(lambda x: attrib_to_feach(x, feach, attrib))
```

```python
def unique_values (col_name, dataframe):
    l = []
    for i in dataframe[col_name]:
        l+=list(i)
    return set(l)
```

```python
def count_of_values (value, l):
    count = 0
    for i in l:
        if i == value:
            count += 1
    return count
```

```python
def make_feature (col_name, dataframe):
    new_feature = unique_values(col_name, dataframe)
    for feature in new_feature:
        dataframe[col_name+'_'+feature] = dataframe[col_name].apply(lambda x: count_of_values(feature, x))
```

```python
col_name = 'user_state'
make_feature(col_name, dataframe)
dataframe.drop(col_name, axis =1, inplace = True)
```

```python
def str_float (l):
    L = []
    if len(l) == 0:
        L.append(0)
    else:
        for i in l:
            L.append(float(i))
    return L
    
```
